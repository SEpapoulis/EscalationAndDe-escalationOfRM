```python
%load_ext rpy2.ipython

from src.pipeline import Comprehensive_Scan

import os
import pandas as pd
import numpy as np
from multiprocessing import Pool , Manager
from src.datastructures import Fasta, ORG
from src.RMdefinitions import MTaseDomains
```

```python
RM = Fasta('data/search_results/ElementsFound.fasta')
```

```python
domains = set()
MTases = Fasta()
assemblies = set()
for el in RM:
    evidence = el.split()[1].split('=')[-1]
    GCF = el.split()[2].split('=')[-1]
    if 'M.' == evidence [0:2]:
        MTases[el] = RM[el]
        assemblies.add(GCF)
    elif set(evidence.split(';')).intersection(MTaseDomains):
        MTases[el] = RM[el]
        assemblies.add(GCF)
MTases.write_file('data/MTases.fasta')
```

```python
def pick16S(fasta):
    size=0
    largest_seq=''
    for el in fasta:
        if 'product=16S' in el:
            if len(fasta[el]) > size:
                largest_seq=el
                size = len(fasta[el])
    return(largest_seq)

GCF_16S = Fasta()
gbdir ='/media/zinserlab/4TbWesterDigital/Prokaryotes/'
files=os.listdir(gbdir)
gcf_file = {file.split('.')[0]:gbdir+file for file in files}

fasta16S = Fasta()
itot = len(assemblies)
for i,assembly in enumerate(assemblies):
    print("{}/{}".format(str(i+1),str(itot)),end="\r")
    try:
        org=ORG(gcf_file[assembly])
        orgfasta=org.get_16Sfasta()
        org16S = pick16S(orgfasta)
        fasta16S['>'+assembly] = orgfasta[org16S]
    except KeyError:
        print('please download '+assembly )
fasta16S.write_file('16S.fasta')
    
```
