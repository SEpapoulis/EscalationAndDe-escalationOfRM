```python
gbdir ='/media/zinserlab/4TbWesterDigital/Prokaryotes/'
archaea=2157
bacteria=2
#last downloaded 7-24-2018
catalog.download_subtree(bacteria,gbdir,genbank=False)
catalog.download_subtree(archaea,gbdir,genbank=False)
```

```python
missing=[]
found=[]
found_dat={}
duplicate=[]
for assembly in gcf:
    try:
        taxid = catalog[assembly].taxid
        phylum=catalog.find_rankname(taxid,'phylum')
        genus=catalog.find_rankname(taxid,'genus')
        name=catalog.get_name(assembly)
        strain=catalog.get_strain(assembly)[0]
        assembly_level = catalog.get_assembly_level(assembly)[0]
        refseq_category = catalog.get_refseq_category(assembly)[0]
        if refseq_category == 'na':
            refseq_category=''
        ftp=catalog.get_ftp(assembly)[0]
        if assembly in found_dat:
            duplicate.append(assembly)
        found_dat[assembly] = [assembly, taxid, phylum,genus,name,strain.split('=')[-1],
                               assembly_level,refseq_category,ftp]
        found.append(assembly)
    except KeyError:
        missing.append(assembly)
        
print(len(found))
print(len(missing))
print(len(duplicate))
```

```python
orgs = pd.DataFrame([found_dat[el] for el in found_dat])
orgs.columns=['assembly','NCBI_taxid','phylum','genus','name','strain','assembly_level','refseq_category','ftp']
orgs.to_csv('data/assembly_details.csv',index=False)
```
